```python
def getcount(sender):
    dummy = msgx[msgx['Sender']==sender]
    return np.sum(dummy['ReplyCount'])

        
def getmsgkeys(senderlist):
    msg_map = {}
    for sender in senderlist:
        msg_map[sender] = getcount(sender)

    msg_map = dict(list(dict(sorted(msg_map.items(),key = lambda x:x[1],reverse=True)).items())[:5])

    return list(msg_map.keys())


def getfinaldict(name):
    
    dummy = msgx[msgx['Sender']==name]
    replylist = list(dummy['Replier'])
    replycount = list(dummy['ReplyCount'])
    maindict = defaultdict(list)
    for i in range(len(replylist)):
        maindict[replylist[i]].append(replycount[i])

    ans = dict(maindict)
    for key,value in ans.items():
        ans[key] = sum(value)

    ans = dict(list(dict(sorted(ans.items(),key = lambda x:x[1],reverse=True)).items())[:5])
    return ans

    
senderlist = []
sender = list(msgx['Sender'])
for i in range(len(sender)):
    if sender[i] not in senderlist:
        senderlist.append(sender[i])

        
## get the msg keys

msgkeys = getmsgkeys(senderlist)
finalmap = {}
for msg in msgkeys:
    finalmap[msg] = getfinaldict(msg)
    
    
finalmap

        
```

```python
def framedf(name,mapper):
    
    dummy = pd.DataFrame(mapper.items(),columns=['User','Count'])
    dummy['Sender'] = [name]*(dummy.shape[0])
    
    return dummy

concatlist = []
for key,value in finalmap.items():
    concatlist.append(framedf(key,value))
```

```python
finaldf = pd.concat(concatlist)
finaldf = finaldf[['Sender','User','Count']]
finaldf.columns = ['Sender','Replier','Count']
finaldf
```
