**First make a map of the person and the message which they are sending,then check out at which indicies the person is repeating,then find the difference count wrt to adjacent index if the diff is >1 that means there was a reply to that persons message,now the difference between current index and adjacent index is the number of replies on average**

```python
message_list = list(df_2018_new['Message'])
person_list = list(df_2018_new['Person'])
top10person = persons_map
print(len(message_list),len(person_list))
```

```python
def getmsgcount(message_list,person_list,person):
    
    idx_count = []
    idx_map = {}
    for i,j in zip(range(len(person_list)),range(len(message_list))):
        if persons_list[i]==person:
            idx_map[i] = message_list[j]
            idx_count.append(i)


    uselist = []
    for i in range(len(idx_count)-1):
        if abs(idx_count[i]-idx_count[i+1])>1:
            uselist.append(idx_count[i])



    msgcount = []
    for i in range(len(uselist)-1):
        msgcount.append(abs(uselist[i]-uselist[i+1]))

    
    return msgcount
```

```python
df_2018_new = df_2018[(df_2018['Month']=='Aug')|(df_2018['Month']=='Sep')|(df_2018['Month']=='Oct')
                     |(df_2018['Month']=='Nov')|(df_2018['Month']=='Dec')]

persons_list = list(df_2018_new['Person'])
persons_map = dict(zip(persons_list,map(persons_list.count,persons_list)))
persons_map = dict(sorted(persons_map.items(),key=lambda x:x[1],reverse=True))
persons_map = dict(list(persons_map.items())[:10])
top10person = list(persons_map.keys())
topmonth = list(df_2018_new['Month'].unique())

msgs = []
final_map = defaultdict(list)
for person in top10person:
    for month in topmonth:
        dummy = df_2018_new[df_2018_new['Month']==month]
        message_list = list(dummy['Message'])
        person_list = list(dummy['Person'])
        final_map[person].append(getmsgcount(message_list,person_list,person))
```

```python
touselist = {}
for key,val in final_map.items():
    dummy = final_map[key]
    count = 0
    for litem in dummy:
        if len(litem)>0:
            count+=1
            
    if count==5:
        touselist[key] = val

touselist
```

```python
finalmapper = defaultdict(list)
for key,val in touselist.items():
    
    dummylist = touselist[key]
    
    for listitem in dummylist:
        include = []
        if len(listitem)==1:
            include.append(listitem[0])
        
        else:   
            for item in range(len(listitem)-1):
                if abs(listitem[item]-listitem[item+1])<=40:
                    include.append(listitem[item])
        
        sumlist = []
        if len(include)==1:
            sumlist.append(include[0])
        else:        
            for inc in range(len(include)-1):
                sumlist.append(abs(include[inc]-include[inc+1]))

        finalmapper[key].append(sum(sumlist))
```

```python
def getdf(mapper,year):
    dummy = pd.DataFrame(mapper.items(),columns = ['User','Reply Count'])
    dummy['Year'] = [year]*dummy.shape[0]
    return dummy

mapperlist = [mapper_2018,mapper_2019,mapper_2020,mapper_2021]
yearlist = [2018,2019,2020,2021]

dflist = []
for mappers,year in zip(mapperlist,yearlist):
    dflist.append(getdf(mappers,year))
    

finaldf =  pd.concat(dflist)
finaldf
```
